```python
import pandas as pd 
tab=pd.read_csv('Données bonheur.csv', sep=';', decimal=",", index_col='Country name')
#on écarte la liste des années
l=list(tab)[1:]
#on normalise toutes les valeurs entre 0 et 1
for name in l :
    tab1=tab[tab[name].notna()]
    min=tab1[name].min()
    max=tab1[name].max()
    ecart=max-min
    tab[name]=(tab[name]-min)/ecart
```

```python
#Création indice bonhuer en faisant la moyenne de tous les paramètres 
tab['Indice bonheur']=tab[l].mean(axis=1)
resultat=tab[['year','Indice bonheur']]
resultat
```
